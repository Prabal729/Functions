4.  Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def prime(a):
  c=0
  for i in range(2,a//2):
    if a%i==0:
      c+=1
  if c==0:
    print(a,end=",")
def check():
  for i in range(1,200):
    prime(i)
check()
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power(a):
  for i in range(a):
    yield 2**i
for i in power(10):
  print(i)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_string_line_by_line(text):
    for line in text.splitlines():
        yield line

# Example usage:
text = """Hello, world!
This is a test.
Reading line by line."""

for line in read_string_line_by_line(text):
    print(line)  # Process each line
```

11.  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
Order Number	Book Title and Author			Quantity	Price per Item
34587		Learning Python,Mark Lutz		4  		40.95
98762		Programming Python,Mark Lutz		5  		56.80
77226		Head First Python,Paul Berry		3  		32.95
88112		Einfuhrung in Python3,Bernd klein	3  		24.99
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
print("Order Number\tBook Title and Author\t\t\tQuantity\tPrice per Item\n34587\t\tLearning Python,Mark Lutz\t\t4  \t\t40.95\n98762\t\tProgramming Python,Mark Lutz\t\t5  \t\t56.80\n77226\t\tHead First Python,Paul Berry\t\t3  \t\t32.95\n88112\t\tEinfuhrung in Python3,Bernd klein\t3  \t\t24.99")
```

```python
orders = [[34587, 40.95, 4], [98762, 56.80, 5], [77226, 32.95, 3], [88112, 24.99, 3]]

result = list(map(lambda x: (x[0], x[1] * x[2] if x[1] * x[2] >= 100 else x[1] * x[2] + 10), orders))

print(result)
```
